```python
# Filter dataset for LicenseIssuedYear between 2015 and 2023
licenses_per_year = df[df["LicenseIssuedYear"].between(2015, 2023)]

# Count the number of issued licenses per year
licenses_per_year = licenses_per_year.groupby("LicenseIssuedYear").size().reset_index(name="Count")

# Save the file in the current working directory
licenses_per_year.to_csv("licenses_issued_2015_2023.csv", index=False)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("NYC_Dog_Licensing_Dataset_20250210.csv", low_memory=False)

# Count the number of licenses per ZIP code
zip_counts = df['ZipCode'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(12, 6))
zip_counts.plot(kind='bar')

# Labels and title
plt.xlabel("ZIP Code")
plt.ylabel("Number of Dog Licenses")
plt.title("Total Number of Dog Licenses per ZIP Code")
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
# Create a DataFrame with ZIP code and their respective dog license counts
zip_counts_df = zip_counts.reset_index()
zip_counts_df.columns = ['ZipCode', 'LicenseCount']

# Export to CSV
zip_counts_df.to_csv("dog_licenses_per_zipcode.csv", index=False)

print("CSV file has been saved as 'dog_licenses_per_zipcode.csv'")
```
